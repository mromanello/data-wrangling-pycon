```python
import requests
import re
```

```python
import urllib3.contrib.pyopenssl
urllib3.contrib.pyopenssl.inject_into_urllib3()
```

```python
def get_story(story_id):
    url = 'https://hacker-news.firebaseio.com/v0/item/%d.json' % story_id
    resp = requests.get(url)
    return resp.json()
```

```python
def get_top_stories():
    url = 'https://hacker-news.firebaseio.com/v0/topstories.json'
    resp = requests.get(url)
    all_stories = [get_story(sid) for sid in resp.json()[:10]]
    return all_stories
```

```python
#top_stories = pd.read_json("../../data/hn.json")
```

```python
top_stories = get_top_stories()
```

```python
MATCHING = (
    ('Python', '(p|P)ython'),
    ('Ruby', '(r|R)uby'),
    ('JavaScript', 'js|(J|j)ava(s|S)cript'),
    ('NodeJS', 'node(\.?)(?:\js|JS)'),
    ('Java', '(j|J)ava[^(S|s)cript]'),
    ('Objective-C', 'Obj(ective?)(?:\ |-)(C|c)'),
    ('Go', '(g|G)o'),
    ('C++',  '(c|C)(\+)+')
)
```

```python
def count_languages():
    stories = get_top_stories()
    final_tallies = {}
    for s in stories:
        long_string = u'{} {}'.format(s.get('title'), s.get('url'))
        for language, regex in dict(MATCHING).items():
            if re.search(regex, long_string):
                if language not in final_tallies.keys():
                    final_tallies[language] = {
                        'score': s.get('score'),
                        'descendants': s.get('descendants')}
                else:
                    final_tallies[language]['score'] += s.get('score')
                    final_tallies[language][
                        'descendants'] += s.get('descendants')
    return final_tallies
```

## working with messy data (# 5)

`applymap` goes cell by cell in a dataframe

```python
import pandas as pd
import numpy as np
from datetime import datetime
```

```python
weather = pd.read_csv('../../data/berlin_weather_oldest.csv')
```

first cleaning step: converting `-9999` values into `np.nan`

```python
weather = weather.applymap(lambda x: np.nan if x == -9999 else x)
```

```python
weather.info
```

```python
# `map` is used when altering value by column, `apply` when by row 
weather['DATE'] = weather['DATE'].map(lambda x: datetime.strptime(str(x), '%Y%m%d').date())
```

```python
weather.notnull().head()
```

```python
weather.shape
```

```python
weather.dropna(how='all', axis=1)
```

```python
weather.dropna(thresh=weather.shape[0] * .1, axis=1) # uses a threshold to drop rows with missing data
```

```python
weather = weather.set_index(pd.DatetimeIndex(weather['DATE']))
```

```python
weather.head()
```

```python
weather.groupby('STATION_NAME').head()
```

```python
weather.groupby('STATION_NAME').resample('D').mean() # filling missing data for some months
```
